```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2
import matplotlib.patches as mpatches
%matplotlib inline
import numpy as np
```

```python
from sklearn import svm, datasets
```

```python
iris = datasets.load_iris()
print(iris)
```

```python
X = iris.data[:, :]

print(X)
```

```python
y = iris.target
print(y)
```

```python
X_scaled = StandardScaler().fit_transform(X)


print(X_scaled)
```

```python
svd = TruncatedSVD(n_components=2, n_iter=500)

print(svd)
```

```python

Y_fitted = svd.fit_transform(X_scaled)


print(Y_fitted)
```

```python

for labels, columns in zip((0, 1, 2),('red', 'green', 'blue')):
    plt.scatter(Y_fitted[y==labels, 0], Y_fitted[y==labels, 1], label=labels, c=columns, marker='.', s=50)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

red_patch = mpatches.Patch(color='red', label='Iris-setosa')
green_patch = mpatches.Patch(color='green', label='Iris-versicolor')
blue_patch = mpatches.Patch(color='blue', label='Iris-virginica')

plt.legend(handles=[red_patch, green_patch, blue_patch])

plt.title("SVD - Iris Data", fontsize=20)
plt.show()
```

```python
def make_meshgrid(x, y, h=.02):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy
```

```python
def plot_contours(ax, clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out

```

```python
C = 1.0  # SVM regularization parameter
models = (svm.SVC(kernel='linear', C=C),
          svm.LinearSVC(C=C))
models = (clf.fit(Y_fitted, y) for clf in models)
```

```python
titles = ('SVC with on Transofrmed data',
          'LinearSVC on Transofrmed data')
```

```python
# Set-up 1x2 grid for plotting.
fig, sub = plt2.subplots(1, 2)
plt2.subplots_adjust(wspace=0.4, hspace=0.4)

X0, X1 = Y_fitted[:, 0], Y_fitted[:, 1]
xx, yy = make_meshgrid(X0, X1)

for clf, title, ax in zip(models, titles, sub.flatten()):
    plot_contours(ax, clf, xx, yy,
                  cmap=plt2.cm.coolwarm, alpha=0.8)
    ax.scatter(X0, X1, c=y, cmap=plt2.cm.coolwarm, s=20, edgecolors='k')
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xlabel('Principle Comp 1')
    ax.set_ylabel('Principle Comp 2')
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)

plt2.show()
```
